```python
from ucimlrepo import fetch_ucirepo
import pandas as pd

# Fetch dataset
iris = fetch_ucirepo(id=53)

# Data (as pandas DataFrames)
X = iris.data.features
y = iris.data.targets

# Combine X (features) and y (targets) into a single DataFrame
#reference: https://pandas.pydata.org/docs/dev/user_guide/merging.html
iris_combined = pd.concat([X, y], axis=1)

# Save the combined dataset to a CSV file
iris_combined.to_csv('files/iris_combined.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#select specific variable
#https://saturncloud.io/blog/how-to-select-specific-csv-columns-using-python-and-pandas/#:~:text=Once%20you%20have%20loaded%20the,column%20names%20as%20the%20argument.

data = pd.read_csv('files/iris_combined.csv')
df = data['class']

#https://datascienceparichay.com/article/create-a-pie-chart-of-pandas-series-values/
#https://sparkbyexamples.com/pandas/how-to-add-plot-legends-in-pandas/#:~:text=Add%20Legend%20Outside%20of%20the%20Plot%20in%20Pandas&text=legend%20using%20plt.-,legend()%20function.,legend%20outside%20of%20the%20plot.
#https://proclusacademy.com/blog/customize_matplotlib_piechart/#slice-colors   

df.value_counts().plot(kind='pie',autopct='%1.0f%%', startangle=90, labeldistance=1.1)


#https://stackoverflow.com/questions/18962063/matplotlib-setting-title-bold-while-using-times-new-roman 
plt.title('CLASS PERCENTAGE',fontweight="bold")
plt.legend(loc='best', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

Now we will represent each of the other four variables _'sepal length' - 'sepal width' - 'petal length' - 'petal width'_ based on the flower _'class'_.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

#Create Histogram
datafile = pd.read_csv('files/iris_combined.csv')

#plotting sepal length with flower class
#reference: https://seaborn.pydata.org/generated/seaborn.stripplot.html#   
#use dodge to separate each flower class and change x for hue. We changed the previous script because, even though the plots were working as intended, we kept receiving this warning - FutureWarning: Passing `palette` without assigning `hue` is deprecated
#reference: https://github.com/mwaskom/seaborn/issues/3450 

#because the next plots would include the plots before,it's ncessary to include $ plt.clf() right after each plot: https://www.activestate.com/resources/quick-reads/how-to-clear-a-plot-in-python/
#subplot reference:https://python-course.eu/numerical-programming/creating-subplots-in-matplotlib.php#:~:text=Activating%20all%20subplot%20of%20the%202x2%20grid%20looks%20like%20this%3A
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1) 

sns.stripplot(data = datafile, y = 'sepal length', hue = 'class', palette='Blues',legend='auto', dodge=True)
plt.title('Sepal length by Flower class')
#plt.savefig("Sepal length per Flower class")
#plt.clf()

plt.subplot(2, 2, 2)  # Create subplot for sepal width
#follow the same steps with the other three variables
sns.stripplot(data = datafile, y = 'sepal width', hue = 'class', palette='Blues',legend='auto', dodge=True)
plt.title('Sepal width by Flower class')
#plt.savefig('Sepal width by Flower class')
#plt.clf()


plt.subplot(2, 2, 3)  # Create subplot for sepal width
sns.stripplot(data = datafile, y = 'petal length', hue = 'class', palette='Blues',legend='auto', dodge=True)
plt.title('Petal length by Flower class')
#plt.savefig('Petal length by Flower class')
#plt.clf()


plt.subplot(2, 2, 4)  # Create subplot for sepal width
sns.stripplot(data = datafile, y = 'petal width', hue = 'class', palette='Blues',legend='auto', dodge=True)
plt.title('Petal width by Flower class')
#plt.savefig('Petal width by Flower class')
#plt.clf()

#plots were displayed as a column 
#reference:https://stackoverflow.com/questions/20057260/how-to-remove-gaps-between-subplots
plt.tight_layout() 
plt.show()


```

>As we mentioned before, the sizes of the flower variables are sometimes overlaping, it would be interesting to have an overview of the variables collectively to determine the most prevalent sizes among the three species.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

datafile = pd.read_csv('files/iris_combined.csv')

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1) 
sns.violinplot(data=datafile['sepal length'])
plt.title('Sepal Length')

plt.subplot(2, 2, 2) 
sns.violinplot(data=datafile['sepal width'])
plt.title('Sepal Width')

plt.subplot(2, 2, 3)  
sns.violinplot(data=datafile['petal length'])
plt.title('Petal Length')

plt.subplot(2, 2, 4)  
sns.violinplot(data=datafile['petal width'])
plt.title('Petal Width')

plt.tight_layout()  
plt.show()


```

>Looking at the plot, the average measurements for 'sepal width' appear quite similar across the species.

>Yet, the variations in petal length and petal width demonstrate distinct patterns, likely due to the unique small size of Iris-setosa, leading to two peaks or fluctuations.

>From this representation we can see that, the average meassure range for sepal length is between 5cm and 6.5cm.

Scatter Plot Maximum/Minimum: In this plot we will analyse the maximum and minimum of each variable depending on the flower type.

```python
import matplotlib.pyplot as plt 
import seaborn as sns

datafile = pd.read_csv('files/iris_combined.csv')
datafile.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'flower class']


# Group data by flower class and compute min and max values for petal length
# By using goupby() function 'flower class', the result will be assigned to the function datatable: $ datatable = dataframe.groupby('flower class')[['sepal length'].agg(['min', 'max'])
# References: https://sparkbyexamples.com/pandas/pandas-group-dataframe-rows-list-groupby/#:~:text=You%20can%20group%20DataFrame%20rows,the%20list%20for%20every%20group. https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html https://sparkbyexamples.com/pandas/pandas-groupby-explained-with-examples/
datatable = datafile.groupby('flower class')['petal length'].agg(['min', 'max'])

# Create subplots with 1 row and 1 column
# Reference: https://stackoverflow.com/questions/34162443/why-do-many-examples-use-fig-ax-plt-subplots-in-matplotlib-pyplot-python
fig, ax = plt.subplots(1,1)
# Scatter plot the min and max values for each flower class
# Now we can define the plot. We want a scatterplot, the reason why we will use the Pandas function scatter(). We used index in order to indicate that we will be sorting data by the index, that is 'flower class' rows (assigned right before through groupby()): $ ax.scatter(datatable.index, datatable['min'], color='blue') $ ax.scatter(datatable.index, datatable['max'], color='yellow')
# Reference: https://sparkbyexamples.com/pandas/get-row-number-in-pandas/#:~:text=You%20can%20get%20the%20row,can%20use%20the%20len(df. https://sparkbyexamples.com/pandas/pandas-groupby-explained-with-examples/
ax.scatter(datatable.index, datatable['min'], color='blue')
ax.scatter(datatable.index, datatable['max'], color='yellow')
ax.set_xlabel('Flower Class max/min')
ax.set_ylabel('Petal Length')
plt.suptitle('Petal Length',weight='bold')


datatable = datafile.groupby('flower class')['sepal length'].agg(['min', 'max'])
fig, ax = plt.subplots(1,1)
ax.scatter(datatable.index, datatable['min'], color='blue')
ax.scatter(datatable.index, datatable['max'], color='yellow')
ax.set_xlabel('Flower Class')
ax.set_ylabel('sepal length')
plt.suptitle('Sepal Length',weight='bold')


datatable = datafile.groupby('flower class')['sepal width'].agg(['min', 'max'])
fig, ax = plt.subplots(1,1)
ax.scatter(datatable.index, datatable['min'], color='blue')
ax.scatter(datatable.index, datatable['max'], color='yellow')
ax.set_xlabel('Flower Class')
ax.set_ylabel('sepal width')
plt.suptitle('Sepal Width',weight='bold')


datatable = datafile.groupby('flower class')['petal width'].agg(['min', 'max'])
fig, ax = plt.subplots(1,1)
ax.scatter(datatable.index, datatable['min'], color='blue')
ax.scatter(datatable.index, datatable['max'], color='yellow')
ax.set_xlabel('Flower Class')
ax.set_ylabel('petal width')
plt.suptitle('Petal Width',weight='bold')

plt.tight_layout() 
plt.show()
```
