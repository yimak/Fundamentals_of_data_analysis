```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
datafile = pd.read_csv('iris_combined.csv')
datafile.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Plot sepal length by Flower class
ax[0, 0].plot(data=datafile, y='sepal length', hue='class', palette='Blues', legend='auto', dodge=True)
ax[0, 0].set_title('Sepal length by Flower class')

# Plot sepal width by Flower class
sns.stripplot(data=datafile, y='sepal width', hue='class', palette='Blues', legend='auto', dodge=True, ax=axes[0, 1])
axes[0, 1].set_title('Sepal width by Flower class')

# Plot petal length by Flower class
sns.stripplot(data=datafile, y='petal length', hue='class', palette='Blues', legend='auto', dodge=True, ax=axes[1, 0])
axes[1, 0].set_title('Petal length by Flower class')

# Plot petal width by Flower class
sns.stripplot(data=datafile, y='petal width', hue='class', palette='Blues', legend='auto', dodge=True, ax=axes[1, 1])
axes[1, 1].set_title('Petal width by Flower class')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
