The Collatz conjecture is a famous unsolved problem in mathematics. The problem is to prove that if you start with any positive
integer $x$ and repeatedly apply the function $f(x)$ below, you always get stuck in the repeating sequence 1, 4, 2, 1, 4, 2, . . .

> First we will start by defining x, which will be a positive integer that will be subject to our calculation. We will start by using the keyword $def$ and assigning the parameter $x$ to a function named $f$. Then we will use $if$ $else$ statement, $if$ the remainder $%$ of $x$ divided by 2 is zero, the return will be the result of dividing $//$ the value of $x$ by two, otherwise or $else$ we should multiply by 3 and add 1.  
Reference: https://atlantictu-my.sharepoint.com/personal/ian_mcloughlin_atu_ie/_layouts/15/onedrive.aspx?id=%2Fpersonal%2Fian%5Fmcloughlin%5Fatu%5Fie%2FDocuments%2Fstudent%5Fshares%2Ffundamentals%5Fof%5Fdata%5Fanalysis%2F1%5Fgeneral%2Ft01v10%5Ffund%5Fdata%5Ftask%5Fone%5Fand%5Frepo%2Emkv&parent=%2Fpersonal%2Fian%5Fmcloughlin%5Fatu%5Fie%2FDocuments&parentview=4 

```python
def f(x):
    # If x is even, divide by 2
    if x % 2 == 0:
         return x // 2 
    else:
        return (3 * x) + 1
```

> By defining $x$ on a $if$ $else$ statement we could only calculate one integer, now we will try to calculate the collatz conjecture until we reach the integer 1. The conjecture is that these sequences always reach 1, no matter which positive integer is chosen to start the sequence. So we will use the keyword $def$ and assign the parameter $x$ to a function collatz. Then we will use the command $while$ or while loop. As the name indicates, as long as the parameter $x$ does not equal 1, print the value of $x$ followed by a comma. Then assign that value to $f(x)$ and start all over again until we reach the integer 1.

```python
def collatz(x):    
    while x != 1:
        print(x, end=', ')
        x=f(x)
    print(x)
```

Suppose you are flipping two coins, each with a probability p of giving heads. Plot the entropy of the total number of heads versus p:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#we will use the binomial random distribution 
# n will designate the trials and p the probability, size will be number of times n (that is, 2 coins) with a probability of p to happen
np.random.binomial(2, 0.5, 10)
```

```python
#representation in plot, flipping two fair coins with a 0.5 probability of heads 100 times.
#the possibilities would be; HH, HT or TT.
rand_coins = np.random.binomial(2,0.5,20)
print(rand_coins)

#create empty plot
fig, ax = plt.subplots()

#count unique values 
x,counts = np.unique(rand_coins, return_counts=True)

ax.bar(x,counts);
```

>Here we can see that the chances to get HH (2 Heads) and TT (0 Heads) are 0.25 each and 0.5 for HT & TH (1 Head). The reason for that is that all outcomes with 1 Head, that is [0,1] & [1,0] are added together.

```python
import numpy as np
import matplotlib.pyplot as plt


# Generate random coin flips with a probability of 0.5 for getting heads
rand_coins = np.random.binomial(2,0.5,1)

# If we get twice heads, np.sum(rand_coins) will be equal to 1, that means that np.sum(rand_coins==n) will count the occurincies of n. 
# If we get twice 0 heads, rand_coins will be equal to 0, that is [0,0].
# The third scenario rand_coins will be equal to 1, which could be either [0,1] or [1,0]

#flip two coins 100 times
rand_coins = np.random.binomial(2, 0.5, 1000)


#count how many times each of them appears
#https://note.nkmk.me/en/python-numpy-count/ 
zero_heads = np.count_nonzero(rand_coins == 0)
one_head = np.count_nonzero(rand_coins == 1)
two_heads = np.count_nonzero(rand_coins == 2)


#calculate the percentage of 0, 1 and 2 out of 100 coins 
perc_0_heads = zero_heads / 1000
perc_1_head = one_head / 1000
perc_2_heads = two_heads / 1000

print(perc_0_heads)
print(perc_1_head)
print(perc_2_heads)


# Probabilities for getting 0, 1, and 2 heads is approximately:
perc_0_heads = 0.25
prob_1_head = 0.5
prob_2_heads = 0.25

# Calculate entropies for each percentage
#https://medium.com/@pradeep_thomas/the-answer-is-42-with-a-pinch-of-salt-information-entropy-d48069c91ba2 
entropy_0_heads = -perc_0_heads * np.log2(perc_0_heads)
entropy_1_head = -perc_1_head * np.log2(perc_1_head)
entropy_2_heads = -perc_2_heads * np.log2(perc_2_heads)


print(entropy_0_heads)
print(entropy_1_head)
print(entropy_2_heads)


# Values for plotting
outcomes = ['0 Heads', '1 Head', '2 Heads']
entropy_values = [entropy_0_heads, entropy_1_head, entropy_2_heads]


# Create a bar plot
plt.figure(figsize=(10,6))
plt.bar(outcomes, entropy_values, color='skyblue')
plt.ylabel('Entropy')
plt.show()
```

>From here we can conclude that, although the propbabilities are different, the entropy will be still be the same and that entropy an probability are two different concepts.
